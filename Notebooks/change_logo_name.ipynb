```python
import os

# Path to the folder containing logo images
folder_path = "/Users/rdhir/Documents/NBAWinPredictions/NBA"

# Mapping of full team names (lowercase filenames) to abbreviations
team_abbr = {
    "hawks": "ATL",
    "celtics": "BOS",
    "nets": "BKN",
    "hornets": "CHA",
    "bulls": "CHI",
    "cavaliers": "CLE",
    "mavericks": "DAL",
    "nuggets": "DEN",
    "pistons": "DET",
    "warriors": "GSW",
    "rockets": "HOU",
    "pacers": "IND",
    "clippers": "LAC",
    "lakers": "LAL",
    "grizzlies": "MEM",
    "heat": "MIA",
    "bucks": "MIL",
    "timberwolves": "MIN",
    "pelicans": "NOP",
    "knicks": "NYK",
    "thunder": "OKC",
    "magic": "ORL",
    "76ers": "PHI",
    "suns": "PHO",
    "trailblazers": "POR",
    "kings": "SAC",
    "spurs": "SAS",
    "raptors": "TOR",
    "jazz": "UTA",
    "wizards": "WAS"
}

# Loop through files and rename if they match
for filename in os.listdir(folder_path):
    name, ext = os.path.splitext(filename)
    key = name.lower()

    if key in team_abbr:
        new_name = f"{team_abbr[key]}.png"
        src = os.path.join(folder_path, filename)
        dst = os.path.join(folder_path, new_name)
        os.rename(src, dst)
        print(f"Renamed: {filename} → {new_name}")
    else:
        print(f"Skipped: {filename}")
```
